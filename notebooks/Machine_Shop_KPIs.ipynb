```python
import pandas as pd #for data manipulation
import matplotlib.pyplot as plt #for visualization
import seaborn as sns #for advanced visualizations

file_path = "/workspaces/codespaces-jupyter/data/Machine_Shop_KPI_Data.csv"
data = pd.read_csv(file_path)

data.head() # Displays the first few rows of the dataset
data.info() #Check the data types and any missing values
#data.describe() # get basic statistics of numeric columns

data = data.dropna() # Dropping rows with missing values (if applicable)

data['Date'] = pd.to_datetime(data['Date']) # Convert 'Date' column to datetime format

data['Machine_Utilization_Rate'] = data['Operating_Hours'] / data['Total_Hours_Available']

plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Date', y='Machine_Utilization_Rate', hue='Machine_ID')
plt.title('Machine Utilization Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Utilization Rate')
plt.xticks(rotation=45)
plt.show()
```
